## Dataset Information

The "spam" concept is diverse: advertisements for products/web sites, make money fast schemes, chain letters, pornography...

The SMS Spam Collection is a set of SMS tagged messages that have been collected for SMS Spam research. It contains one set of SMS messages in English of 5,574 messages, tagged according being ham (legitimate) or spam.

## Attributes

- SMS Messages
- Label (spam/ham)

## Import modules

```python
import pandas as pd
import numpy as np
import nltk
import re
from nltk.corpus import stopwords
```

## Loading the dataset

```python
df = pd.read_csv('/content/spam.csv', encoding='latin1')
df.head()
```

```python
# get necessary columns for processing
df = df[['v2', 'v1']]
# df.rename(columns={'v2': 'messages', 'v1': 'label'}, inplace=True)
df = df.rename(columns={'v2': 'messages', 'v1': 'label'})
df.head()
```

## Preprocessing the dataset

```python
# check for null values
df.isnull().sum()
```

```python
nltk.download('stopwords')
STOPWORDS = set(stopwords.words('english'))

def clean_text(text):
    # convert to lowercase
    text = text.lower()
    # remove special characters
    text = re.sub(r'[^0-9a-zA-Z]', ' ', text)
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text)
    # remove stopwords
    text = " ".join(word for word in text.split() if word not in STOPWORDS)
    return text
```

```python
# clean the messages
df['clean_text'] = df['messages'].apply(clean_text)
df.head()
```

## Input Split

```python
X = df['clean_text']
y = df['label']
```

## Model Training

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer

def classify(model, X, y):
    # train test split
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, shuffle=True, stratify=y)
    # model training
    pipeline_model = Pipeline([('vect', CountVectorizer()),
                              ('tfidf', TfidfTransformer()),
                              ('clf', model)])
    pipeline_model.fit(x_train, y_train)

    print('Accuracy:', pipeline_model.score(x_test, y_test)*100)

#     cv_score = cross_val_score(model, X, y, cv=5)
#     print("CV Score:", np.mean(cv_score)*100)
    y_pred = pipeline_model.predict(x_test)
    print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
classify(model, X, y)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
classify(model, X, y)
```

```python
from sklearn.svm import SVC
model = SVC(C=3)
classify(model, X, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
classify(model, X, y)
```

## Implement user input and prediction

### Subtask:
Create a new cell that re-initializes and trains the best performing model (SVC) within a pipeline, takes text input from the user, cleans the text using the `clean_text` function, and then uses the trained model to predict and print whether the input is 'ham' or 'spam'.

```python
# Re-initialize and train the SVC model within a pipeline
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.svm import SVC

pipeline_svc = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', SVC(C=3)) # SVC with C=3 was the best performing model
])
pipeline_svc.fit(X, y)

# Function to predict user input
def predict_spam_ham(text):
    clean = clean_text(text)
    prediction = pipeline_svc.predict([clean])
    print(f"The message is: {prediction[0]}")

# Take user input
user_message = input("Enter an SMS message to classify: ")
predict_spam_ham(user_message)
```
